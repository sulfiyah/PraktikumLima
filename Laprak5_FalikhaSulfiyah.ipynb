```python
year = int(input("Enter a year: "))

if (year % 4) == 0:
    if (year % 100) == 0:
        if (year % 400) == 0:
            print(year, "adalah tahun kabisat.")
        else:
            print(year, "bukan tahun kabisat.")
    else:
        print(year, "adalah tahun kabisat.")
else:
    print(year, "bukan tahun kabisat.")
```
